3.	Create an 1D array of 100 evenly spaced numbers between 0 and 100 using NumPy and save it to variable X.
a.	Create a plot using plt.subplots() and plot X versus X^2
b.	Create a scatter plot of X versus the exponential of X (np.exp(X))

```python
X= np.arange(100)
X2=np.array(X**2)
expX=np.exp(X)
fig,ax1= plt.subplots()
plt.title('x vs. x**2')
ax1.set_xlabel('x',color='red',size='large')
ax1.set_ylabel('x**2',color='red',size='large')
ax1.plot(X,X2)
fig,ax2= plt.subplots()
plt.title('x vs. exp(x)')
ax2.set_xlabel('x',color='red',size='large')
ax2.set_ylabel('exp(x)',color='red',size='large')
ax2.plot(X,expX)
plt.show()
```

5.	Remove all the symbols (\$\,)from the 'Price' column and convert  and update it to  64-bit int numbers using astype(‘int64’) Hint. Look for string functions like replace/strip). The data frame will be look like after that:

```python
car_sales['Price']=car_sales['Price'].str.replace('$','').str.replace(',','').astype('int64')
car_sales.head()
```

6.	Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column look up the cumsum() function.

```python
car_sales['Total Sales']=np.cumsum(car_sales['Price'])
car_sales.head()
```

7.	Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today). Check out “date_range” function and keep the periods=len(car_sales). The data frame will be look like that:


```python
car_sales['Sale Date']=pd.date_range(dt.today(), periods=len(car_sales))
car_sales.head()
```

8.	Use the matplotlib line plot function to plot the 'Price' column versus the 'Total Sales' column (Give proper title and appropriate labels.

```python
fig,ax5= plt.subplots()
plt.title('Price vs. Total Sales line plot')
ax5.set_xlabel('Price',color='black',size='large')
ax5.set_ylabel('Total Sales',color='black',size='large')
ax5.plot(car_sales['Price'],car_sales['Total Sales'])
```

9.	Create a matplotlib scatter plot of the 'Odometer (KM)' vs 'Price' column.

```python
fig,ax6=plt.subplots()
plt.title('Odometer vs. Price scatter plot')
ax6.set_xlabel('Odometer (KM)',color='black',size='large')
ax6.set_ylabel('Price',color='black',size='large')
ax6.scatter(car_sales['Odometer (KM)'],car_sales['Price'])
plt.show()
```


16.	Find out Pearson correlation between 'Total Sales','Price','Odometer (KM)'. Comment on the correlation 

```python
df=pd.DataFrame(car_sales,columns=['Total Sales','Price','Odometer (KM)'])
#car_sales['Total Sales']['Price']['Odometer (KM)']
sns.heatmap(df.corr(),annot=True,annot_kws={'size':30},linewidth=1)
plt.show()
```

```python
matrix=car_sales.drop(['Make','Colour','Doors','Sale Date'],axis=1)
matrix.corr()
sns.heatmap(matrix.corr(),annot=True, annot_kws={'size': 30},linewidth=1)
plt.show()
#The correlation between odometer and price is a relatively strong inverse correlation while the correlation between price and total sales is a relatively weak direct correlation.
```
